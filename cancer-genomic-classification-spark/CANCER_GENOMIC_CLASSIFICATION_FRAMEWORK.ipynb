```python
# =============================================================================
# CANCER GENOMIC CLASSIFICATION FRAMEWORK
# Advanced Machine Learning Pipeline with Ensemble Methods
# =============================================================================

# Environment configuration for distributed computing
!apt-get update -qq && apt-get install -y openjdk-8-jdk-headless > /dev/null
!pip install pyspark==3.5.0 findspark plotly==5.15.0

import os
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-8-openjdk-amd64"

import findspark
findspark.init()

# Core computational libraries
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from plotly.subplots import make_subplots

# Spark ML components
from pyspark.sql import SparkSession, DataFrame
from pyspark.ml.classification import (DecisionTreeClassifier,
                                     RandomForestClassifier,
                                     GBTClassifier)
from pyspark.ml.evaluation import (BinaryClassificationEvaluator,
                                 MulticlassClassificationEvaluator)
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml import Pipeline, Transformer
from pyspark.sql.functions import udf, col, when
from pyspark.sql.types import DoubleType, ArrayType
from pyspark.ml.linalg import Vectors, VectorUDT

print("üîß Computational environment configured successfully!")

# =============================================================================
# SPARK CLUSTER INITIALIZATION
# =============================================================================

def initialize_computing_engine():
    """Configure distributed computing environment with optimized settings"""
    computing_engine = SparkSession.builder \
        .appName("GenomicAnalysisEngine") \
        .config("spark.driver.memory", "2g") \
        .config("spark.executor.memory", "2g") \
        .config("spark.sql.adaptive.enabled", "true") \
        .config("spark.sql.adaptive.coalescePartitions.enabled", "true") \
        .config("spark.serializer", "org.apache.spark.serializer.KryoSerializer") \
        .getOrCreate()

    computing_engine.sparkContext.setLogLevel("ERROR")

    engine_info = computing_engine.version
    print(f"üöÄ Distributed computing engine initialized: Spark {engine_info}")
    return computing_engine

# Launch computing engine
computing_engine = initialize_computing_engine()

# =============================================================================
# SYNTHETIC GENOMIC DATA GENERATOR
# =============================================================================

class GenomicDataSimulator:
    """Simulates genomic expression data with biological patterns"""

    def __init__(self, random_state=123):
        self.random_state = random_state
        np.random.seed(random_state)

    def generate_expression_profiles(self, n_observations=800, n_genes=150):
        """Create synthetic gene expression profiles with cancer signatures"""

        expression_data = []
        diagnostic_labels = []

        for observation_idx in range(n_observations):
            # Simulate baseline gene expression
            base_expression = np.random.normal(loc=0.0, scale=1.2, size=n_genes)

            # Introduce cancer-specific expression patterns
            cancer_risk_markers = np.concatenate([
                np.random.normal(0.8, 0.6, 40),   # Up-regulated in cancer
                np.random.normal(-0.5, 0.4, 35),  # Down-regulated in cancer
                np.random.normal(0.0, 1.0, 75)    # Neutral genes
            ])

            # Combine patterns
            final_expression = base_expression + cancer_risk_markers

            # Determine diagnostic outcome
            cancer_probability = (np.tanh(np.mean(final_expression[:40]) * 2) + 1) / 2
            has_cancer = 1 if cancer_probability > 0.45 else 0

            # Add realistic noise
            if np.random.random() < 0.08:
                has_cancer = 1 - has_cancer

            expression_data.append(Vectors.dense(final_expression.astype(np.float64)))
            diagnostic_labels.append(float(has_cancer))

        # Create structured dataset
        genomic_df = computing_engine.createDataFrame(
            zip(diagnostic_labels, expression_data),
            ["diagnosis", "genomic_features"]
        )

        return genomic_df

# Generate synthetic genomic dataset
print("üß¨ Creating synthetic genomic expression dataset...")
data_simulator = GenomicDataSimulator()
genomic_dataset = data_simulator.generate_expression_profiles(800, 150)
print("‚úÖ Genomic dataset successfully generated!")

# =============================================================================
# EXPLORATORY GENOMIC ANALYSIS
# =============================================================================

def perform_genomic_analysis(dataset):
    """Comprehensive analysis of genomic dataset characteristics"""

    print("\n" + "üîç GENOMIC DATASET ANALYSIS REPORT" + "="*40)

    # Basic dataset metrics
    total_observations = dataset.count()
    feature_dimension = len(dataset.first().genomic_features)

    print(f"üìä Dataset Scale: {total_observations} observations")
    print(f"üß¨ Genomic Features: {feature_dimension} genes")

    # Diagnostic distribution
    diagnosis_distribution = dataset.groupBy("diagnosis").count().orderBy("diagnosis")
    print("\nüè• Diagnostic Distribution:")
    diagnosis_distribution.show()

    # Sample inspection
    print("\nüìã Data Sample Preview:")
    dataset.select("diagnosis", "genomic_features").show(3, truncate=True)

    # Convert for visualization
    sample_for_viz = dataset.limit(150).toPandas()

    # Create comprehensive visualizations
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))

    # Diagnostic distribution
    diagnosis_counts = dataset.groupBy("diagnosis").count().toPandas()
    axes[0,0].bar(['Healthy', 'Cancer'], diagnosis_counts['count'],
                 color=['lightgreen', 'lightcoral'], alpha=0.7)
    axes[0,0].set_title('Diagnostic Category Distribution', fontweight='bold')
    axes[0,0].set_ylabel('Patient Count')

    # Expression value distribution
    expression_values = sample_for_viz['genomic_features'].apply(lambda x: x.toArray().mean())
    axes[0,1].hist(expression_values, bins=25, color='skyblue', alpha=0.7, edgecolor='black')
    axes[0,1].set_title('Average Gene Expression Distribution', fontweight='bold')
    axes[0,1].set_xlabel('Mean Expression Level')
    axes[0,1].set_ylabel('Frequency')

    # Feature correlation heatmap (first 8 genes)
    expression_matrix = np.array([x.toArray() for x in sample_for_viz['genomic_features']])
    correlation_matrix = np.corrcoef(expression_matrix[:, :8], rowvar=False)
    sns.heatmap(correlation_matrix, ax=axes[1,0], cmap='RdBu_r', center=0,
                xticklabels=[f'Gene_{i+1}' for i in range(8)],
                yticklabels=[f'Gene_{i+1}' for i in range(8)])
    axes[1,0].set_title('Gene Expression Correlation Matrix', fontweight='bold')

    # Expression by diagnosis
    healthy_expr = [x.toArray().mean() for x in sample_for_viz[sample_for_viz['diagnosis']==0]['genomic_features']]
    cancer_expr = [x.toArray().mean() for x in sample_for_viz[sample_for_viz['diagnosis']==1]['genomic_features']]
    axes[1,1].boxplot([healthy_expr, cancer_expr], labels=['Healthy', 'Cancer'])
    axes[1,1].set_title('Expression Levels by Diagnosis', fontweight='bold')
    axes[1,1].set_ylabel('Mean Expression Level')

    plt.tight_layout()
    plt.show()

    return sample_for_viz

# Execute genomic analysis
genomic_sample = perform_genomic_analysis(genomic_dataset)

# =============================================================================
# ADVANCED ML PIPELINE ARCHITECTURE
# =============================================================================

class GenomicClassifierEnsemble:
    """Ensemble of classifiers for genomic data analysis"""

    def __init__(self, computing_engine):
        self.engine = computing_engine
        self.trained_classifiers = {}
        self.performance_metrics = {}
        self.validation_results = {}

    def partition_dataset(self, dataset, validation_ratio=0.25):
        """Split dataset into training and validation subsets"""
        training_subset, validation_subset = dataset.randomSplit(
            [1-validation_ratio, validation_ratio], seed=42
        )

        print(f"üìö Training samples: {training_subset.count()}")
        print(f"üß™ Validation samples: {validation_subset.count()}")

        return training_subset, validation_subset

    def configure_tree_classifier(self):
        """Configure decision tree with hyperparameter grid"""
        tree_model = DecisionTreeClassifier(
            labelCol="diagnosis",
            featuresCol="genomic_features",
            seed=42
        )

        hyperparameter_grid = ParamGridBuilder() \
            .addGrid(tree_model.maxDepth, [4, 8, 12]) \
            .addGrid(tree_model.minInstancesPerNode, [1, 5]) \
            .addGrid(tree_model.impurity, ["gini", "entropy"]) \
            .build()

        return tree_model, hyperparameter_grid

    def configure_forest_classifier(self):
        """Configure random forest ensemble"""
        forest_model = RandomForestClassifier(
            labelCol="diagnosis",
            featuresCol="genomic_features",
            featureSubsetStrategy="sqrt",
            seed=42
        )

        hyperparameter_grid = ParamGridBuilder() \
            .addGrid(forest_model.numTrees, [25, 50, 75]) \
            .addGrid(forest_model.maxDepth, [6, 10]) \
            .addGrid(forest_model.subsamplingRate, [0.8, 1.0]) \
            .build()

        return forest_model, hyperparameter_grid

    def configure_gradient_boosting(self):
        """Configure gradient boosting classifier"""
        gbt_model = GBTClassifier(
            labelCol="diagnosis",
            featuresCol="genomic_features",
            seed=42
        )

        hyperparameter_grid = ParamGridBuilder() \
            .addGrid(gbt_model.maxDepth, [3, 5]) \
            .addGrid(gbt_model.maxIter, [20, 40]) \
            .addGrid(gbt_model.stepSize, [0.05, 0.1]) \
            .build()

        return gbt_model, hyperparameter_grid

    def assess_classifier_performance(self, trained_model, test_data, classifier_name):
        """Comprehensive performance evaluation"""

        # Generate predictions
        prediction_results = trained_model.transform(test_data)

        # Configure evaluation metrics
        auc_evaluator = BinaryClassificationEvaluator(
            labelCol="diagnosis",
            rawPredictionCol="rawPrediction",
            metricName="areaUnderROC"
        )

        accuracy_evaluator = MulticlassClassificationEvaluator(
            labelCol="diagnosis",
            predictionCol="prediction",
            metricName="accuracy"
        )

        f1_evaluator = MulticlassClassificationEvaluator(
            labelCol="diagnosis",
            predictionCol="prediction",
            metricName="f1"
        )

        # Calculate performance metrics
        auc_score = auc_evaluator.evaluate(prediction_results)
        accuracy_score = accuracy_evaluator.evaluate(prediction_results)
        f1_score = f1_evaluator.evaluate(prediction_results)

        performance_summary = {
            'ROC_AUC': auc_score,
            'Classification_Accuracy': accuracy_score,
            'F1_Score': f1_score
        }

        print(f"\nüìà {classifier_name.upper()} PERFORMANCE:")
        for metric_name, metric_value in performance_summary.items():
            print(f"   {metric_name}: {metric_value:.4f}")

        # Store results
        self.performance_metrics[classifier_name] = {
            'metrics': performance_summary,
            'predictions': prediction_results
        }

        return prediction_results, performance_summary

    def train_classifier_ensemble(self, training_data, validation_data):
        """Train multiple classifiers with cross-validation"""

        print("\nüöÄ INITIATING CLASSIFIER TRAINING" + "="*30)

        # Define classifier configurations
        classifier_configurations = {
            'decision_tree': self.configure_tree_classifier(),
            'random_forest': self.configure_forest_classifier(),
            'gradient_boosting': self.configure_gradient_boosting()
        }

        # Cross-validation setup
        cv_evaluator = BinaryClassificationEvaluator(
            labelCol="diagnosis",
            rawPredictionCol="rawPrediction",
            metricName="areaUnderROC"
        )

        training_results = {}

        for clf_name, (clf_model, param_grid) in classifier_configurations.items():
            print(f"\n--- Training {clf_name.replace('_', ' ').title()} ---")

            # Cross-validation training
            cross_validator = CrossValidator(
                estimator=clf_model,
                estimatorParamMaps=param_grid,
                evaluator=cv_evaluator,
                numFolds=4,
                seed=42,
                parallelism=2
            )

            # Model training
            cv_trained_model = cross_validator.fit(training_data)
            self.trained_classifiers[clf_name] = cv_trained_model

            # Performance evaluation
            predictions, metrics = self.assess_classifier_performance(
                cv_trained_model, validation_data, clf_name
            )
            training_results[clf_name] = metrics

            # Feature importance analysis
            try:
                if hasattr(cv_trained_model.bestModel, 'featureImportances'):
                    feature_weights = cv_trained_model.bestModel.featureImportances
                    print(f"   üîç Feature importance analysis completed")
            except AttributeError:
                print(f"   ‚ö†Ô∏è  Feature importance not available")

        return training_results

    def visualize_performance_comparison(self):
        """Create performance comparison visualizations"""

        if not self.performance_metrics:
            print("No performance data available for visualization")
            return None

        # Prepare performance data
        performance_data = pd.DataFrame({
            clf_name: metrics['metrics']
            for clf_name, metrics in self.performance_metrics.items()
        }).T

        # Create visualization
        plt.figure(figsize=(14, 8))

        performance_data.plot(kind='bar', ax=plt.gca(),
                           color=['#2E86AB', '#A23B72', '#F18F01'])
        plt.title('Genomic Classifier Performance Comparison', fontsize=14, fontweight='bold')
        plt.ylabel('Performance Score', fontsize=12)
        plt.xlabel('Machine Learning Models', fontsize=12)
        plt.xticks(rotation=45, ha='right')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.grid(axis='y', alpha=0.3)
        plt.tight_layout()
        plt.show()

        return performance_data

# Initialize and execute the ensemble pipeline
print("\n" + "üß† INITIALIZING GENOMIC CLASSIFICATION PIPELINE" + "="*20)
genomic_pipeline = GenomicClassifierEnsemble(computing_engine)
train_set, validation_set = genomic_pipeline.partition_dataset(genomic_dataset)
ensemble_results = genomic_pipeline.train_classifier_ensemble(train_set, validation_set)

# =============================================================================
# PERFORMANCE ANALYSIS AND VISUALIZATION
# =============================================================================

# Generate performance comparison
performance_comparison = genomic_pipeline.visualize_performance_comparison()

# Detailed results analysis
print("\n" + "üèÜ FINAL MODEL PERFORMANCE RANKING" + "="*40)

for model_name, model_metrics in ensemble_results.items():
    print(f"\n{model_name.upper().replace('_', ' ')}:")
    for metric, value in model_metrics.items():
        print(f"  {metric}: {value:.4f}")

# Identify top-performing model
best_performing_model = max(ensemble_results.items(), key=lambda x: x[1]['ROC_AUC'])
print(f"\nüéñÔ∏è  TOP PERFORMING MODEL: {best_performing_model[0].replace('_', ' ').title()}")
print(f"üìä Best ROC-AUC Score: {best_performing_model[1]['ROC_AUC']:.4f}")

# =============================================================================
# FEATURE IMPORTANCE ANALYSIS
# =============================================================================

def analyze_genomic_feature_importance(pipeline, top_features=12):
    """Analyze and visualize important genomic features"""

    print("\n" + "üß¨ GENOMIC FEATURE IMPORTANCE ANALYSIS" + "="*30)

    for model_name, trained_model in pipeline.trained_classifiers.items():
        try:
            if hasattr(trained_model.bestModel, 'featureImportances'):
                importance_scores = trained_model.bestModel.featureImportances

                # Extract top important features
                feature_importance_pairs = [(idx, score) for idx, score in enumerate(importance_scores)]
                top_important_features = sorted(feature_importance_pairs,
                                             key=lambda x: x[1], reverse=True)[:top_features]

                print(f"\nüîù Top {top_features} Features for {model_name.replace('_', ' ').title()}:")
                for rank, (feature_index, importance_score) in enumerate(top_important_features, 1):
                    print(f"   {rank:2d}. Genomic Marker {feature_index:4d}: {importance_score:.4f}")

                # Visualize feature importance
                plt.figure(figsize=(10, 6))
                features, scores = zip(*top_important_features)
                plt.barh(range(len(features)), scores[::-1],
                        color=plt.cm.viridis(np.linspace(0, 1, len(features))))
                plt.yticks(range(len(features)), [f'Gene_{f}' for f in features[::-1]])
                plt.xlabel('Feature Importance Score', fontsize=12)
                plt.title(f'Top Genomic Features - {model_name.replace("_", " ").title()}',
                         fontweight='bold')
                plt.tight_layout()
                plt.show()

        except Exception as error:
            print(f"‚ö†Ô∏è  Could not analyze features for {model_name}: {error}")

# Execute feature importance analysis
analyze_genomic_feature_importance(genomic_pipeline)

# =============================================================================
# PREDICTION DEMONSTRATION
# =============================================================================

def demonstrate_predictions(pipeline, validation_data, n_examples=8):
    """Show prediction examples with confidence scores"""

    best_model_identifier = max(pipeline.performance_metrics.items(),
                             key=lambda x: x[1]['metrics']['ROC_AUC'])[0]
    best_predictions = pipeline.performance_metrics[best_model_identifier]['predictions']

    print(f"\nüîÆ PREDICTION SAMPLES (Best Model: {best_model_identifier}):")
    print("="*65)

    # Extract prediction probabilities
    probability_extractor = udf(lambda prob_vector: float(prob_vector[1]), DoubleType())

    prediction_samples = best_predictions.select(
        "diagnosis",
        "prediction",
        probability_extractor("probability").alias("malignancy_probability")
    ).limit(n_examples).toPandas()

    # Display formatted predictions
    for sample_idx, sample_row in prediction_samples.iterrows():
        actual_diagnosis = "Malignant" if sample_row['diagnosis'] == 1 else "Benign"
        predicted_diagnosis = "Malignant" if sample_row['prediction'] == 1 else "Benign"
        prediction_confidence = (sample_row['malignancy_probability']
                              if sample_row['prediction'] == 1
                              else 1 - sample_row['malignancy_probability'])
        correctness_indicator = "‚úÖ" if sample_row['diagnosis'] == sample_row['prediction'] else "‚ùå"

        print(f"Sample {sample_idx+1:2d}: "
              f"Actual: {actual_diagnosis:>9} | "
              f"Predicted: {predicted_diagnosis:>9} | "
              f"Confidence: {prediction_confidence:.3f} {correctness_indicator}")

# Show prediction demonstrations
demonstrate_predictions(genomic_pipeline, validation_set)

# =============================================================================
# COMPREHENSIVE EXECUTION SUMMARY
# =============================================================================

print("\n" + "üìã PIPELINE EXECUTION SUMMARY" + "="*45)

total_observations = genomic_dataset.count()
training_observations = train_set.count()
validation_observations = validation_set.count()

print(f"üìä Dataset Overview:")
print(f"   ‚Ä¢ Total genomic profiles: {total_observations}")
print(f"   ‚Ä¢ Training observations: {training_observations} ({training_observations/total_observations*100:.1f}%)")
print(f"   ‚Ä¢ Validation observations: {validation_observations} ({validation_observations/total_observations*100:.1f}%)")

print(f"\nü§ñ Trained Classifiers: {len(genomic_pipeline.trained_classifiers)}")
for model_name, metrics in ensemble_results.items():
    print(f"   ‚Ä¢ {model_name.replace('_', ' ').title()}: AUC = {metrics['ROC_AUC']:.4f}")

best_classifier = max(ensemble_results.items(), key=lambda x: x[1]['ROC_AUC'])
print(f"\nüèÖ Optimal Classifier Performance:")
print(f"   ‚Ä¢ Algorithm: {best_classifier[0].replace('_', ' ').title()}")
print(f"   ‚Ä¢ ROC-AUC Score: {best_classifier[1]['ROC_AUC']:.4f}")
print(f"   ‚Ä¢ Accuracy: {best_classifier[1]['Classification_Accuracy']:.4f}")
print(f"   ‚Ä¢ F1-Score: {best_classifier[1]['F1_Score']:.4f}")

print(f"\nüéâ Genomic classification pipeline completed successfully!")

# Cleanup computing resources
# computing_engine.stop()
```
